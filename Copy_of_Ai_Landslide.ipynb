what is the landslide risk leve given environmental conditions

```python
state = 'safe'
# -- define the rules
def landslide_rule(rainfall, slope, soil):
    if rainfall > 100 and slope > 30 :
      if soil == 'clay':
        return 'High Risk'
      else: # If rainfall and slope are high, but soil is not clay
        return 'Medium Risk'
    elif rainfall > 50:
      return 'Low Risk'
    else:
      return 'Very Low Risk'
```

```python
#score the land
def landslide_score(rainfall, earthquakes, slope):
    score=0
    if rainfall > 100:
        score +=2 #score=score+2
    if earthquakes > 2:
        score +=2
    if slope > 30:
        score +=2
    if score >= 5:
        return 'High Risk'
    elif score >= 3:
        return 'Medium Risk'
    else:
        return 'Low Risk'
```

```python
#update the state
def update_state(rainfall):
  global state
  if rainfall > 100:
    state = 'Danger'
  else:
    state = 'Safe'
  return state
```

```python
#make a chatbot
def chatbot():
  print('Hello, I am a chatbot. How can I help you today?')
  while True:
    user_input = input('You : ').lower()

    if user_input == 'exit' or user_input == 'stop':
      print('Goodbye!')
      break

    elif 'risk' in user_input:
      rainfall = float(input('How much rainfall is expected? '))
      slope = float(input('What is the slope of the area? '))
      soil = input('What is the soil type? (clay/sand):')
      earthquakes= int(input('How many earthquakes have occurred in the last 30 days? '))

      rule = landslide_rule(rainfall, slope, soil)
      score = landslide_score(rainfall, earthquakes, slope)
      state = update_state(rainfall)

      print(f'AI(rule):{rule}')
      print(f'AI(score):{score}')
      print(f'AI(state):{state}')

    elif 'hello' in user_input:
      print(f'AI: Hello! Ask me about landslide risk.')
    else:
      print(f'AI: I don\'t understand. Try "risk".')
```
